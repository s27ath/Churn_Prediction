Identification of input and output variables:  
X='gender','tenure', 'InternetService', 'OnlineSecurity','OnlineBackup',
 'DeviceProtection','TechSupport','Contract','PaperlessBilling',
  'MonthlyCharges','TotalCharges'  
y='churn'

its a supervised learning classification problem

accuracy is the evaluation metric



```python
#input variables
X=df[['gender','tenure', 'InternetService', 'OnlineSecurity','OnlineBackup','DeviceProtection','TechSupport','Contract','PaperlessBilling','PaymentMethod','MonthlyCharges','TotalCharges']]
#output Variable
y=df['Churn']
```

Splitting the data into train and test

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,train_size=0.70,random_state=100)
```

```python
print(X_train.shape,y_train.shape)
print(X_test.shape,y_test.shape)
```

Data preparation on num and cat train data







```python
X_train_cat=X_train.select_dtypes(include=['object'])
X_train_num=X_train.select_dtypes(include=['int64','float64'])
```

```python
X_train_num.head()
```

```python
X_train_cat.head()
```

Applying standerdization on num columns

```python
from sklearn.preprocessing import StandardScaler
```

```python
Scaler=StandardScaler()
```

```python
X_train_num_rescaled=pd.DataFrame(Scaler.fit_transform(X_train_num),
                                  columns=X_train_num.columns,
                                  index=X_train_num.index)
```

```python
X_train_num_rescaled.head()
```

```python
X_train_num_rescaled.shape
```

Applying one hot encoding on cat columns

```python
X_train_cat['InternetService'].value_counts(normalize=True)
```

```python
X_train_cat['Contract'].value_counts(normalize=True)
```

```python
X_train_cat['PaymentMethod'].value_counts(normalize=True)
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
encoder=OneHotEncoder(drop='first',sparse = False)
```

```python
X_train_cat_ohe=pd.DataFrame(encoder.fit_transform(X_train_cat),
                             columns=encoder.get_feature_names_out(X_train_cat.columns),
                             index=X_train_cat.index)
```

```python
X_train_cat_ohe.head()
```

```python
X_train_cat_ohe.shape
```

```python
X_train_transformed=pd.concat([X_train_num_rescaled,X_train_cat_ohe],axis=1)
```

```python
X_train_transformed.head()
```

```python
X_train_transformed.shape
```

data prep on test data

```python
X_test_num=X_test.select_dtypes(include=['int64','float64'])
X_test_cat=X_test.select_dtypes(include=['object'])
```

```python
X_test_num.head()
```

```python
X_test_num_rescaled=pd.DataFrame(Scaler.fit_transform(X_test_num),
                                  columns=X_test_num.columns,
                                  index=X_test_num.index)
```

```python
X_test_num_rescaled.head()
```

```python
X_test_num_rescaled.shape
```

Onehotencoding on test data

```python
X_test_cat_ohe=pd.DataFrame(encoder.fit_transform(X_test_cat),
                             columns=encoder.get_feature_names_out(X_test_cat.columns),
                             index=X_test_cat.index)
```

```python
X_test_cat_ohe.head()
```

```python
X_test_transformed=pd.concat([X_test_num_rescaled,X_test_cat_ohe],axis=1)
```

```python
X_test_transformed.head()
```

```python
X_test_transformed.shape
```

Model training,prediction and evaluation

KNN Classifier

```python
from sklearn.neighbors import KNeighborsClassifier
classifier=KNeighborsClassifier()
classifier.fit(X_train_transformed,y_train)
```

```python
y_test_pred = classifier.predict(X_test_transformed)
```

```python
from sklearn import metrics
print("Accuracy of KNN classifier model: ",metrics.accuracy_score(y_test, y_test_pred))
```

```python
temp_df = pd.DataFrame({'Actual': y_test, 'Predicted': y_test_pred})

temp_df.head()
```

```python
from sklearn.linear_model import LogisticRegression
classifier=LogisticRegression()
classifier.fit(X_train_transformed,y_train)
```

```python
y_test_pred = classifier.predict(X_test_transformed)
```

```python
print("Accuracy of Logistic Regression classifier model: ",metrics.accuracy_score(y_test, y_test_pred))
```

```python
cm = metrics.confusion_matrix(y_test, y_test_pred)

print("Confusion Matrix:")
print(cm)
```

```python
tp = cm[1][1]
tn = cm[0][0]
fp = cm[0][1]
fn = cm[1][0]

accuracy = (tp+tn)/(tp+tn+fp+fn)
print(f'accuracy: {accuracy}')

cm = metrics.confusion_matrix(y_test, y_test_pred)

fig = metrics.ConfusionMatrixDisplay(cm, display_labels=classifier.classes_)

fig.plot()
```

```python
from sklearn.tree import DecisionTreeClassifier
classifier=DecisionTreeClassifier()
classifier.fit(X_train_transformed,y_train)
```

```python
y_test_pred = classifier.predict(X_test_transformed)
```

```python
print("Accuracy of Decision Tree classifier model: ",metrics.accuracy_score(y_test, y_test_pred))
```

```python
cm = metrics.confusion_matrix(y_test, y_test_pred)

print("Confusion Matrix:")
print(cm)
```

```python
cm = metrics.confusion_matrix(y_test, y_test_pred)

fig = metrics.ConfusionMatrixDisplay(cm, display_labels=classifier.classes_)

fig.plot()
```

```python
from sklearn.ensemble import RandomForestClassifier
classifier=RandomForestClassifier()
classifier.fit(X_train_transformed,y_train)
```

```python
y_test_pred = classifier.predict(X_test_transformed)
```

```python
print("Accuracy of Random Forest classifier model: ",metrics.accuracy_score(y_test, y_test_pred))
```

```python
cm = metrics.confusion_matrix(y_test, y_test_pred)

print("Confusion Matrix:")
print(cm)
```

```python
cm = metrics.confusion_matrix(y_test, y_test_pred)

fig = metrics.ConfusionMatrixDisplay(cm, display_labels=classifier.classes_)

fig.plot()
```

```python
from sklearn import svm
classifier = svm.SVC()
classifier.fit(X_train_transformed, y_train)
```

```python
y_test_pred = classifier.predict(X_test_transformed)
```

```python
print("Accuracy of Support Vector Machine classifier model: ",metrics.accuracy_score(y_test, y_test_pred))
```

```python
cm = metrics.confusion_matrix(y_test, y_test_pred)

print("Confusion Matrix:")
print(cm)
```

```python
cm = metrics.confusion_matrix(y_test, y_test_pred)

fig = metrics.ConfusionMatrixDisplay(cm, display_labels=classifier.classes_)

fig.plot()
```
